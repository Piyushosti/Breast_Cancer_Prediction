## Making seperate dcolums from the Dataframe 

```python
X=dataframe.drop(columns='diagnosis')
Y=dataframe['diagnosis']
```

```python
X
```

```python
Y
```

## Converting M and B into the form of 0 ans 1
  M for Maligant(0)
  
  B for Balign (1)
  

```python
Y = Y.replace({'M': 0, 'B': 1})
diagnosis_counts = Y.value_counts()
print(diagnosis_counts)
```

```python
sns.countplot(x=dataframe['diagnosis'])
plt.title('diagnosis')
```

## Separting for Training and Test Data 


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

## Model Training

Logistic Regression 


```python
logistic_reg_model = LogisticRegression()
```

## Fitting the Model


```python
logistic_reg_model.fit(X_train, Y_train)
```

## Evalute the model 

```python
X_train_prediciton=logistic_reg_model.predict(X_train)
```

```python
traing_data_accuracy = accuracy_score(Y_train, X_train_prediciton)
```

```python
print(traing_data_accuracy)
```

## Evalute For the Test Data

```python
X_test_prediciton=logistic_reg_model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediciton)
```

```python
print(test_data_accuracy)
```
